def manual_adaboost(M, df, X_val):

    df = df.copy()

    # Initialize validation prediction
    y_val_pred = np.zeros(len(X_val))
    sum_of_alphas=0
    

    for m in range(M):

        
        # Train weak learner
       
        X = df.iloc[:, :8]
        y = df['target']

        dt = DecisionTreeRegressor(max_depth=1, random_state=42)
        dt.fit(X, y)

        
        # Training predictions
       
        df['y_pred'] = dt.predict(X)

        
        # Normalized error
       
        error = np.abs(y - df['y_pred'])
        error = error / error.max()
        df['error'] = error
        errors = df['error']

        
        # Your weighted error E
        #0.00097->initial weight E= sigma(error*weight) we are taking initial weight in each newly form dataframe to start with as we are
        #already upsampling so it ultimately leading to E=wht*(sigma(error))
        
        E = errors.sum() * 0.00097
        E = np.clip(E, 1e-10, 0.999)

       
        # Your alpha
        
        alpha1 = alpha(E)
        sum_of_alphas += alpha1

      
        # Update validation prediction
       
        y_val_pred += alpha1 * dt.predict(X_val)
        

       
        # Update weights 
      
        df['updated_wht'] = df.apply(lambda row: updated_wht(row, alpha1), axis=1)

        # Normalize weights
        df['updated_wht'] /= df['updated_wht'].sum()

      
        # Resampling 
       
        df['cumsum_upper'] = df['updated_wht'].cumsum()
        df['cumsum_lower'] = df['cumsum_upper'] - df['updated_wht']

        indices = create_df(df)
        df = df.iloc[indices,:]

        
    y_val_pred /= sum_of_alphas

    return y_val_pred

import numpy as np
from sklearn.tree import DecisionTreeRegressor

def manual_adaboost(M, df, X_val):
    df = df.copy()
    
   
    if 'updated_wht' not in df.columns:
        df['updated_wht'] = 1 / len(df)

   
    predictors = []
    alphas = []

    for m in range(M):
       
        X = df.iloc[:, :8] 
        y = df['target']
        
        # Train on current weighted distribution
        dt = DecisionTreeRegressor(max_depth=1, random_state=42)
        dt.fit(X, y)
        

        # 2. Calculate Loss/Error
       
        df['y_pred'] = dt.predict(X)
        
        
        error = np.abs(y - df['y_pred'])
        
        # Normalize error
        max_error = error.max()
        if max_error == 0: max_error = 1e-10
        df['error'] = error / max_error
        
        # Calculate Average Loss (L_bar)
        # Since  resampled, the weights are implicitly 1/N
        L_bar = df['error'].mean()
        

        # 3. Stopping Condition & Alpha
       
        # If model is worse than random guessing, stop or discard
        if L_bar >= 0.5:
            break
            
        # Beta calculation
        beta = L_bar / (1 - L_bar)
        
        # Alpha 
        if beta == 0: beta = 1e-10
        alpha = np.log(1 / beta)
        
        predictors.append(dt)
        alphas.append(alpha)

       
        # 4. Update Weights 
      
        # Decrease weight for good predictions (low error)
        # Increase weight for bad predictions (high error)
        # Using the standard R2 update rule: w * beta^(1-error)
        df['updated_wht'] *= np.power(beta, (1 - df['error']))
        
        # Normalize 
        df['updated_wht'] /= df['updated_wht'].sum()

       
        # 5. Resample (Fast Way)
       
        indices = np.random.choice(len(df), size=len(df), replace=True, p=df['updated_wht'])
        df = df.iloc[indices,:].reset_index(drop=True)

  
    # 6. Final Prediction: Weighted Median

   
  
    preds = np.array([model.predict(X_val) for model in predictors]).T
    
    # We need to find the weighted median for every row
    final_predictions = []
    model_weights = np.array(alphas)
    
   
    for i in range(len(preds)):
        row_preds = preds[i]
        
        # Sort predictions
        sorted_idx = np.argsort(row_preds)
        sorted_preds = row_preds[sorted_idx]
        sorted_weights = model_weights[sorted_idx]
        
        # Find the point where cumulative weight hits 50%
        cumulative_weight = np.cumsum(sorted_weights)
        cutoff = cumulative_weight[-1] * 0.5
        
        median_idx = np.searchsorted(cumulative_weight, cutoff)
        final_predictions.append(sorted_preds[median_idx])

    return np.array(final_predictions)

def manual_adaboost(M, df, X_val, lr=0.1):

        y = df['target']
        
       
        # min_samples_leaf=10: Prevents fitting to single outliers
        # max_features='sqrt': Decorrelates trees (forces diversity)
        # criterion='friedman_mse': Often better for boosting
        dt = DecisionTreeRegressor(
            max_depth=1,              
            min_samples_leaf=10,      
            max_features='sqrt',     
            criterion='friedman_mse', 
            random_state=42 + m      
        )
        dt.fit(X, y)
        
        
        df['y_pred'] = dt.predict(X)
        error = np.abs(y - df['y_pred'])
        
        max_error = error.max()
        if max_error == 0: max_error = 1e-10
        df['error'] = error / max_error
        
        L_bar = df['error'].mean()
        
        if L_bar >= 0.5:
            break
            
        beta = L_bar / (1 - L_bar)
        if beta == 0: beta = 1e-10
        
        alpha = np.log(1 / beta)
        alpha *= lr 
        
        predictors.append(dt)
        alphas.append(alpha)

        # Update Weights
        df['updated_wht'] *= np.power(beta, (1 - df['error']) * lr)
        df['updated_wht'] /= df['updated_wht'].sum()

        # Resample
        indices = np.random.choice(len(df), size=len(df), replace=True, p=df['updated_wht'])
        df = df.iloc[indices].reset_index(drop=True)

    
    # Final Prediction
    
    preds = np.array([t.predict(X_val) for t in predictors]).T

<h2>Optimised Model</h2>

```python
def manual_adaboost(M, df,  lr=0.1): 
    df = df.copy()
    X = df.iloc[:, :8]
    y = df['target']
    
    # Initialize weights: 1/N for every row
    # We will UPDATE this array, not the dataframe rows
    sample_weights = np.ones(len(df)) / len(df)

    predictors = []
    alphas = []

    for m in range(M):
      
        dt = DecisionTreeRegressor(max_depth=1, random_state=42)
        dt.fit(X, y, sample_weight=sample_weights)
        
      
        # 2. Calculate Error
       
        y_pred = dt.predict(X)
        error = np.abs(y - y_pred)
        
        # Normalize error
        max_error = error.max()
        if max_error == 0: max_error = 1e-10
        error_norm = error / max_error
        
       
        # We use the sample_weights to calculate the mean error
        L_bar = np.average(error_norm, weights=sample_weights)
        
       
        # 3. Stopping Condition
     
        if L_bar >= 0.5:
            break
            
        beta = L_bar / (1 - L_bar)
        if beta == 0: beta = 1e-10
        
        # Calculate Alpha
        alpha = np.log(1 / beta)
        
        # Apply Learning Rate
        alpha *= lr 
        
        predictors.append(dt)
        alphas.append(alpha)

      
        # 4. Update Weights (No Resampling)
      
        
        # We update the specific weight for each row in place
        sample_weights *= np.power(beta, (1 - error_norm) * lr)
        
        # Normalize weights 
        sample_weights /= sample_weights.sum()

    return alphas,predictors
        

        
```

```python
def prediction(X_test):
    
    
    preds = np.array([t.predict(X_test) for t in predictors]).T
    model_weights = np.array(alphas)
    final_predictions = []

    for i in range(len(preds)):
        row_preds = preds[i]
        
        # Sort predictions
        sorted_idx = np.argsort(row_preds)
        sorted_preds = row_preds[sorted_idx]
        sorted_weights = model_weights[sorted_idx]
        
        #  Weighted Median
        cumsum_weight = np.cumsum(sorted_weights)
        cutoff = cumsum_weight[-1] * 0.5
        median_idx = np.searchsorted(cumsum_weight, cutoff)
        
        final_predictions.append(sorted_preds[median_idx])

    return np.array(final_predictions)
```

```python
M_values = [5, 10, 20, 40, 60, 80, 100, 150, 200, 500, 1000, 5000, 10000]
```

```python
from sklearn.metrics import r2_score

r2_scores = []

for M in M_values:
    print(f"Running AdaBoost for M = {M}")
    # train each model first
    alphas,predictors=manual_adaboost(
        M=M,
        df=df_train,
        
    )
    # predict on that model then
    y_val_pred = prediction(X_val)

    r2 = r2_score(y_val, y_val_pred)
    r2_scores.append(r2)

    print(f"Validation R for M={M}: {r2}\n")
```

<h3>3. Analysis of Differences</h3>
<ul>
    <li><b>Score Similarity:</b> The R scores are nearly identical (0.44 vs 0.45). This confirms that the manual implementation correctly follows the mathematical logic of the AdaBoost.R2 algorithm. The minor difference (approx 0.01) is likely due to:
        <ul>
            <li><b>Floating Point Precision:</b> Sklearn uses highly optimized Cython/C types which handle rounding differently than Python's standard float types.</li>

</ul>

<h3>4. Conclusion</h3>
<p>The manual implementation is logically sound and replicates the behavior of the library version. The fact that both models plateaued at an R of ~0.45 indicates that the limiting factor is the <b>Model Architecture</b> (Decision Stumps with <code>max_depth=1</code>) rather than the boosting algorithm itself. To achieve higher accuracy, deeper trees (e.g., <code>max_depth=4</code>) would be required to capture the complex non-linear relationships observed in the EDA phase.</p>



<h2>Code Evolution: Journey to a Stable AdaBoost Regressor</h2>

<p>Developing the manual AdaBoost implementation required several iterations to stabilize the algorithm and match the performance of the Scikit-Learn library. Below is the summary of the problems encountered and the solutions applied at each stage.</p>

<h3>Version 1: The "Basic" Resampling Approach inspired by adaboost classification</h3>
<ul>
    <li><b>Mechanism:</b> Used <code>np.random.choice</code> to resample the dataset based on weights. For the final prediction, it calculated the <b>Weighted Mean</b> of all trees.</li>
    <li><b>Observation:</b> The R score exploded to negative infinity (e.g., -20,000).</li>
    <li><b>Reason for Failure:</b> In regression, outliers can have massive errors. When using a Weighted Mean, a single bad tree with a huge prediction value can pull the entire average away from the target. AdaBoost.R2 specifically requires a <b>Weighted Median</b> to be robust against these outliers.</li>
</ul>

<h3>Version 2: Fixing the Math (Weighted Median)</h3>
<ul>
    <li><b>Change:</b> Switched the final prediction logic from Mean to <b>Weighted Median</b>.</li>
    <li><b>Observation:</b> The score became realistic (-1.0 to +0.5) but was highly <b>unstable/oscillating</b>. The validation score would jump wildly (e.g., 0.42 &rarr; -0.36 &rarr; 0.08).</li>
    <li><b>Reason for Failure:</b> The model was still using physical resampling (copying rows). Random sampling introduced too much variance; in some iterations, "unlucky" draws caused the weak learner (Decision Stump) to fit to noise rather than the signal, causing the weights to swing violently.</li>
</ul>

<h3>Version 3: Adding Regularization </h3>
<ul>
    <li><b>Change:</b> Added <code>min_samples_leaf=10</code> and <code>max_features='sqrt'</code> to the Decision Tree.</li>
    <li><b>Observation:</b> Stability improved slightly, but performance still crashed at high iterations (M=500 score dropped to -0.86).</li>
    <li><b>Reason for Failure:</b> Even with "handcuffs," the resampling method itself was flawed for high iterations. As weights became extreme, the resampled dataset consisted of just a few unique outliers duplicated hundreds of times (the "Echo Chamber" effect). The model memorized this noise and lost the true pattern.</li>
</ul>

<h3>Version 4: The Final "Reweighting" Solution (Current Code)</h3>
<ul>
    <li><b>Change:</b> Removed <code>np.random.choice</code> entirely. Instead, passed the <code>sample_weight</code> array directly to the <code>DecisionTreeRegressor.fit()</code> method.</li>
    <li><b>Observation:</b> The score became perfectly smooth and stable. It climbed to ~0.45 and plateaued, exactly matching the Scikit-Learn reference implementation.</li>
    <li><b>Conclusion:</b> <b>Reweighting</b> is superior to <b>Resampling</b> for regression because it allows for continuous, precise weight updates without the information loss or noise introduced by random sampling.</li>
</ul>
