```python
def manual_adaboost(M, df, X_val):

    df = df.copy()

    # Initialize validation prediction
    y_val_pred = np.zeros(len(X_val))

    for m in range(M):

        # -----------------------------
        # Train weak learner
        # -----------------------------
        X = df.iloc[:, :8]
        y = df['Concrete compressive strength(MPa, megapascals)']

        dt = DecisionTreeRegressor(max_depth=1, random_state=42)
        dt.fit(X, y)

        # -----------------------------
        # Training predictions
        # -----------------------------
        df['y_pred'] = dt.predict(X)

        # -----------------------------
        # Normalized error
        # -----------------------------
        error = np.abs(y - df['y_pred'])
        error = error / error.max()
        df['error'] = error
        errors = df['error']

        # -----------------------------
        # Your weighted error E
        # -----------------------------
        E = errors.sum() * 0.00097
        E = np.clip(E, 1e-10, 0.999)

        # -----------------------------
        # Your alpha
        # -----------------------------
        alpha1 = alpha(E)

        # -----------------------------
        # Update validation prediction
        # -----------------------------
        y_val_pred += alpha1 * dt.predict(X_val)

        # -----------------------------
        # Update weights (YOUR FUNCTION)
        # -----------------------------
        df['updated_wht'] = df.apply(updated_wht, axis=1)

        # Normalize weights
        df['updated_wht'] /= df['updated_wht'].sum()

        # -----------------------------
        # Resampling (YOUR FUNCTION)
        # -----------------------------
        df['cumsum_upper'] = df['updated_wht'].cumsum()
        df['cumsum_lower'] = df['cumsum_upper'] - df['updated_wht']

        indices = create_df(df)
        df = df.iloc[indices].reset_index(drop=True)

    return y_val_pred
```

import numpy as np
from sklearn.tree import DecisionTreeRegressor

def manual_adaboost(M, df, X_val):
    df = df.copy()
    
    # Initialize weights uniformly if not present
    if 'updated_wht' not in df.columns:
        df['updated_wht'] = 1 / len(df)

    # Lists to store our trained trees and their weights (alphas)
    predictors = []
    alphas = []

    for m in range(M):
        # -----------------------------
        # 1. Train Weak Learner
        # -----------------------------
        X = df.iloc[:, :8]  # Adjust column slicing if needed
        y = df['target']
        
        # Train on current weighted distribution (implicitly handled by resampling)
        dt = DecisionTreeRegressor(max_depth=1, random_state=42)
        dt.fit(X, y)
        
        # -----------------------------
        # 2. Calculate Loss/Error
        # -----------------------------
        df['y_pred'] = dt.predict(X)
        
        # Calculate absolute error
        error = np.abs(y - df['y_pred'])
        
        # Normalize error to 0-1 range (Crucial for AdaBoost.R2)
        max_error = error.max()
        if max_error == 0: max_error = 1e-10
        df['error'] = error / max_error
        
        # Calculate Average Loss (L_bar)
        # Since we resampled, the weights are implicitly 1/N
        L_bar = df['error'].mean()
        
        # -----------------------------
        # 3. Stopping Condition & Alpha
        # -----------------------------
        # If model is worse than random guessing, stop or discard
        if L_bar >= 0.5:
            break
            
        # Beta calculation (standard AdaBoost.R2)
        beta = L_bar / (1 - L_bar)
        
        # Alpha (weight of the tree)
        if beta == 0: beta = 1e-10
        alpha = np.log(1 / beta)
        
        predictors.append(dt)
        alphas.append(alpha)

        # -----------------------------
        # 4. Update Weights (Vectorized)
        # -----------------------------
        # Decrease weight for good predictions (low error)
        # Increase weight for bad predictions (high error)
        # Using the standard R2 update rule: w * beta^(1-error)
        df['updated_wht'] *= np.power(beta, (1 - df['error']))
        
        # Normalize so they sum to 1
        df['updated_wht'] /= df['updated_wht'].sum()

        # -----------------------------
        # 5. Resample (Fast Way)
        # -----------------------------
        # NEW (Correct)
        indices = np.random.choice(len(df), size=len(df), replace=True, p=df['updated_wht'])
        df = df.iloc[indices,:].reset_index(drop=True)

    # -----------------------------
    # 6. Final Prediction: Weighted Median
    # -----------------------------
    # Get predictions from all trees
    # shape: (num_validation_samples, num_trees)
    preds = np.array([model.predict(X_val) for model in predictors]).T
    
    # We need to find the weighted median for every row
    final_predictions = []
    model_weights = np.array(alphas)
    
    # Vectorized Weighted Median is hard, so we loop through rows (fast enough for validation)
    for i in range(len(preds)):
        row_preds = preds[i]
        
        # Sort predictions
        sorted_idx = np.argsort(row_preds)
        sorted_preds = row_preds[sorted_idx]
        sorted_weights = model_weights[sorted_idx]
        
        # Find the point where cumulative weight hits 50%
        cumulative_weight = np.cumsum(sorted_weights)
        cutoff = cumulative_weight[-1] * 0.5
        
        median_idx = np.searchsorted(cumulative_weight, cutoff)
        final_predictions.append(sorted_preds[median_idx])

    return np.array(final_predictions)

import numpy as np
from sklearn.tree import DecisionTreeRegressor

def manual_adaboost(M, df, X_val, lr=0.1): 
    # Safety Check
    df = df.copy().reset_index(drop=True)
    
    if 'updated_wht' not in df.columns:
        df['updated_wht'] = 1 / len(df)

    predictors = []
    alphas = []

    for m in range(M):
        X = df.iloc[:, :8]
        y = df['target']
        
        # -----------------------------
        # IMPROVEMENT 1: ROBUST STUMP
        # -----------------------------
        # min_samples_leaf=10: Prevents fitting to single outliers
        # max_features='sqrt': Decorrelates trees (forces diversity)
        # criterion='friedman_mse': Often better for boosting
        dt = DecisionTreeRegressor(
            max_depth=1,              # STRICTLY 1
            min_samples_leaf=10,      # Regularization
            max_features='sqrt',      # Decorrelation
            criterion='friedman_mse', # Better splitting
            random_state=42 + m       # Change seed slightly for diversity
        )
        dt.fit(X, y)
        
        # -----------------------------
        # Standard AdaBoost Steps
        # -----------------------------
        df['y_pred'] = dt.predict(X)
        error = np.abs(y - df['y_pred'])
        
        max_error = error.max()
        if max_error == 0: max_error = 1e-10
        df['error'] = error / max_error
        
        L_bar = df['error'].mean()
        
        if L_bar >= 0.5:
            break
            
        beta = L_bar / (1 - L_bar)
        if beta == 0: beta = 1e-10
        
        alpha = np.log(1 / beta)
        alpha *= lr 
        
        predictors.append(dt)
        alphas.append(alpha)

        # Update Weights
        df['updated_wht'] *= np.power(beta, (1 - df['error']) * lr)
        df['updated_wht'] /= df['updated_wht'].sum()

        # Resample
        indices = np.random.choice(len(df), size=len(df), replace=True, p=df['updated_wht'])
        df = df.iloc[indices].reset_index(drop=True)

    # -----------------------------
    # Final Prediction: Weighted Median
    # -----------------------------

    model_weights = np.array(alphas)
    final_predictions = []

    for i in range(len(preds)):

        sorted_idx = np.argsort(row_preds)
        sorted_preds = row_preds[sorted_idx]
        sorted_weights = model_weights[sorted_idx]
        
        cumsum_weight = np.cumsum(sorted_weights)
        cutoff = cumsum_weight[-1] * 0.5
        median_idx = np.searchsorted(cumsum_weight, cutoff)
        
        final_predictions.append(sorted_preds[median_idx])

    return np.array(final_predictions)

import numpy as np
from sklearn.tree import DecisionTreeRegressor

def manual_adaboost(M, df, X_val, lr=0.1): # Keeping lr=0.1 as per Sklearn default
    df = df.copy()
    X = df.iloc[:, :8]
    y = df['target']
    
    # Initialize weights: 1/N for every row
    # We will UPDATE this array, not the dataframe rows
    sample_weights = np.ones(len(df)) / len(df)

    predictors = []
    alphas = []

    for m in range(M):
        # -----------------------------
        # 1. Train Weak Learner (Using Weights, NOT Resampling)
        # -----------------------------
        # This is the secret sauce. We pass the weights directly to the tree.
        dt = DecisionTreeRegressor(max_depth=1, random_state=42)
        dt.fit(X, y, sample_weight=sample_weights)
        
        # -----------------------------
        # 2. Calculate Error
        # -----------------------------
        y_pred = dt.predict(X)
        error = np.abs(y - y_pred)
        
        # Normalize error (Linear Loss)
        max_error = error.max()
        if max_error == 0: max_error = 1e-10
        error_norm = error / max_error
        
        # Calculate Average Loss (Weighted Mean of Error)
        # We use the sample_weights to calculate the mean error
        L_bar = np.average(error_norm, weights=sample_weights)
        
        # -----------------------------
        # 3. Stopping Condition
        # -----------------------------
        if L_bar >= 0.5:
            break
            
        beta = L_bar / (1 - L_bar)
        if beta == 0: beta = 1e-10
        
        # Calculate Alpha (Model Weight)
        alpha = np.log(1 / beta)
        
        # Apply Learning Rate
        alpha *= lr 
        
        predictors.append(dt)
        alphas.append(alpha)

        # -----------------------------
        # 4. Update Weights (No Resampling)
        # -----------------------------
        # Update rule: weight * beta^(1-error)
        # We update the specific weight for each row in place
        sample_weights *= np.power(beta, (1 - error_norm) * lr)
        
        # Normalize weights to sum to 1
        sample_weights /= sample_weights.sum()
        
        # NOTE: We do NOT use np.random.choice or create a new df.
        # The 'sample_weights' array carries all the information to the next loop.

    # -----------------------------
    # 5. Final Prediction: Weighted Median
    # -----------------------------
    preds = np.array([t.predict(X_val) for t in predictors]).T
    model_weights = np.array(alphas)
    final_predictions = []

    for i in range(len(preds)):
        row_preds = preds[i]
        
        # Sort predictions
        sorted_idx = np.argsort(row_preds)
        sorted_preds = row_preds[sorted_idx]
        sorted_weights = model_weights[sorted_idx]
        
        # Find Weighted Median
        cumsum_weight = np.cumsum(sorted_weights)
        cutoff = cumsum_weight[-1] * 0.5
        median_idx = np.searchsorted(cumsum_weight, cutoff)
        
        final_predictions.append(sorted_preds[median_idx])

    return np.array(final_predictions)
